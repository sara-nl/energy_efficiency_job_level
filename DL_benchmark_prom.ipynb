```python
import pandas as pd
import numpy as np
from pathlib import Path

from sklearn import set_config
set_config(transform_output = "pandas")
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# folder_path =  Path('./system_analytics_2024/prom_data/benchmark_single_node_job_data')
folder_path = Path('/projects/2/prjs1098/system_analytics_2024/prom_data/benchmark_single_node_job_data')

benchmark_job_file= folder_path / 'benchmark_job_data_cleaned.csv'
benchmark_prom_file= folder_path / 'benchmark_job_data_with_prom.parquet.gzip'

df_benchmark_job = pd.read_csv(benchmark_job_file)
df_benchmark_prom = pd.read_parquet(benchmark_prom_file)
display(df_benchmark_job.head(), df_benchmark_prom.head())
```

```python
"""   
If we need to change the mapping from name to class, here we can also do it.
"""
mapping_name_to_class = {'HPCG_MPIOnly %nnodes=1':'memory_intensive',
                        'Palabos_CPU_WS %nnodes=1 %ppn=full':'memory_intensive',
                         'Cp2k_Cpu %nnodes=1':'high_CPU',
                         'HemePure_CPU_SS_small %nnodes=1 %ppn=full':'memory_intensive',
                        'HemePure_CPU_SS_small %nnodes=1 %ppn=half':'memory_intensive_half',
                        'GROMACS_CPU %n_cores=32':'high_CPU',
                        'GROMACS_GPU %n_gpus=1':'high_CPU_GPU'               
    }
df_benchmark_job['class'] = df_benchmark_job['name'].apply(lambda x: mapping_name_to_class[x])

df_benchmark_job['perfomance_value'] = pd.cut(df_benchmark_job['pref_value'],
                                              bins=[0, 5, 500, 20000], labels=['low', 'medium', 'high'])


display(
        # df_benchmark_job['name'].value_counts(),
        # df_benchmark_job['environment'].value_counts(),
        df_benchmark_job['class'].value_counts(),
        # df_benchmark_job['perfomance_value'].value_counts()
        )

df_benchmark_job.sort_values('job_id', inplace=True)
# Set the distingusher here
distinguisher = 'class'
```

```python
df_benchmark_prom.drop_duplicates(inplace=True)
df_benchmark_prom.sort_values(['job_id', 'timestamp'], inplace=True)
# add a dummy index
df_benchmark_prom['increasing_index'] = df_benchmark_prom['shifted_time'] // 15

# get the maximum length for all the jobs
MAX_JOB_LENGTH = df_benchmark_prom['job_id'].value_counts().max()
max_job_length_index = np.arange(MAX_JOB_LENGTH)

# get the number of jobs and their value
uniqe_job = df_benchmark_prom['job_id'].unique()
NUMBER_OF_UNIQ_JOB = len(uniqe_job)

# create a dummy data frame
two_dim_index = pd.MultiIndex.from_product([uniqe_job, max_job_length_index], names=["job_id", "increasing_index"])
two_dim_index_df = two_dim_index.to_frame(index=False).reset_index(drop=True)
two_dim_index_df.head()
```

```python
# merge the becnhmark data to get a new data frame where length of all jobs are equal
df_benchmark_prom_length_equal = pd.merge(two_dim_index_df, df_benchmark_prom, how='left', on=['job_id', 'increasing_index'])
display(df_benchmark_prom_length_equal.head())
display(df_benchmark_prom_length_equal['job_id'].value_counts())
```
